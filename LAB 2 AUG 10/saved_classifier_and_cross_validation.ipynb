## **Creating Datasets**

```python
import numpy as np
np.random.seed(0)  # make execution reproducible
```

```python
# making a linearly separable dataset
# number of samples = 1000, dimension of input vector = 10

basic_dataset_input_1 = np.random.uniform(-100.0, 100.0, (1000, 11))
basic_dataset_input_1[:, 0] = np.ones(1000)
basic_dataset_ground_truth_weights_1 = np.random.uniform(-5.0, 5.0, 11)
basic_dataset_output_1 = np.matmul(basic_dataset_input_1, basic_dataset_ground_truth_weights_1)
basic_dataset_output_1[basic_dataset_output_1 >= 0.0] = 1
basic_dataset_output_1[basic_dataset_output_1 < 0.0] = 0
print(basic_dataset_input_1.shape)
print(basic_dataset_ground_truth_weights_1.shape)
print(basic_dataset_output_1.shape)
print(basic_dataset_input_1)
print(basic_dataset_ground_truth_weights_1)
print(basic_dataset_output_1)
```

```python
# making a semi-linearly separable dataset
# number of samples = 1000, dimension of input vector = 10

basic_dataset_input_2 = np.random.uniform(-100.0, 100.0, (1000, 11))
basic_dataset_input_2[:, 0] = np.ones(1000)
basic_dataset_ground_truth_weights_2 = np.random.uniform(-5.0, 5.0, 11)
basic_dataset_output_2 = np.matmul(basic_dataset_input_2, basic_dataset_ground_truth_weights_2)
basic_dataset_output_2[basic_dataset_output_2 >= 0.0] = 1
basic_dataset_output_2[basic_dataset_output_2 < 0.0] = 0
basic_dataset_input_2 += np.random.uniform(-5.0, 5.0, (1000, 11))
print(basic_dataset_input_2.shape)
print(basic_dataset_ground_truth_weights_2.shape)
print(basic_dataset_output_2.shape)
print(basic_dataset_input_2)
print(basic_dataset_ground_truth_weights_2)
print(basic_dataset_output_2)
```

```python
# making a non-linearly separable dataset
# number of samples = 1000, dimension of input vector = 10

basic_dataset_input_3 = np.random.uniform(-100.0, 100.0, (1000, 11))
basic_dataset_input_3[:, 0] = np.ones(1000)
basic_dataset_ground_truth_weights_3 = np.random.uniform(-5.0, 5.0, 11)
basic_dataset_output_3 = np.matmul(basic_dataset_input_3, basic_dataset_ground_truth_weights_3)
basic_dataset_output_3[basic_dataset_output_3 >= 0.0] = 1
basic_dataset_output_3[basic_dataset_output_3 < 0.0] = 0
basic_dataset_input_3 = np.random.uniform(-100.0, 100.0, (1000, 11))
print(basic_dataset_input_3.shape)
print(basic_dataset_ground_truth_weights_3.shape)
print(basic_dataset_output_3.shape)
print(basic_dataset_input_3)
print(basic_dataset_ground_truth_weights_3)
print(basic_dataset_output_3)
```

## **Binary Classification, Train Test Split, Cross Validation, Performance Metrics**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

kf = KFold(n_splits=5)

knn_kfold = []
prediction_kfold = []
metrics_kfold = []

for train_index, test_index in kf.split(basic_dataset_input_1): #this loop will runf or 5 times
  X_train, X_test = basic_dataset_input_1[train_index], basic_dataset_input_1[test_index]
  y_train, y_test = basic_dataset_output_1[train_index], basic_dataset_output_1[test_index]

  knn_kfold.append(KNeighborsClassifier(n_neighbors=5))
  knn_kfold[-1].fit(X_train, y_train)
  prediction = knn_kfold[-1].predict(X_test)

  acc = accuracy_score(y_test, prediction)     # TP+TN / (TP + FP + TN + FN)
  prec = precision_score(y_test, prediction)   # TP/TP+FP
  rec = recall_score(y_test, prediction)
  f1 = f1_score(y_test, prediction)

  prediction_kfold.append(prediction)
  metrics_kfold.append([acc, prec, rec, f1])

for metrics in metrics_kfold:
  print(metrics)

#find mean and variance for performance metrics of the models
```
